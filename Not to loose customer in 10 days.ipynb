```python
def replace(dtrain1):
    for column in dtrain1.columns[dtrain1.columns.isin(['joined_through_referral','used_special_discount','offer_application_preference','past_complaint'])]:
          dtrain1[column]=dtrain1[column].replace(('Yes','No'),(1, 0))
    return dtrain1 
```

```python
dtrain2=replace(dtrain1.copy())
```

```python
dtrain2.head()
```

```python
dtrain2['referral_id'].value_counts()
```

```python
dtrain2['points_in_wallet']=dtrain2['points_in_wallet'].fillna(value=dtrain['points_in_wallet'].mean())
```

```python
dtrain2=dtrain2.dropna()
```

```python
dtrain2.isnull().any()
```

```python
dtrain2.head()
```

```python
def encode_data(dtrain2):
    for col in dtrain2.columns[dtrain2.columns.isin(['gender','security_no','region_category','membership_category','referral_id','preferred_offer_types','medium_of_operation','internet_option','complaint_status','feedback'])]:
        dtrain2[col]=dtrain2[col].factorize()[0]
    return dtrain2    
```

```python
dtrain3=encode_data(dtrain2)
```

```python
dtrain3.head()
```

```python
dtrain3['joining_day']=pd.to_datetime(dtrain3.joining_date,format="%Y-%m-%d").dt.day
dtrain3['joining_month']=pd.to_datetime(dtrain3.joining_date,format="%Y-%m-%d").dt.day
dtrain3['joining_year']=pd.to_datetime(dtrain3.joining_date,format="%Y-%m-%d").dt.year
```

```python
dtrain3=dtrain3.drop('joining_date',axis=1)
```
